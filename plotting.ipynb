Just change the path of the `file_train` and `file_val` to the one you want to generate plot for 

```python
import numpy as np
import matplotlib.pyplot as plt 
```

### Losses

```python
# training
file_train = open('/scratch/network/hishimwe/losses/train/train_30epochs','r')
loss_train = file_train.readlines()
loss_train = np.array(loss_train) 
loss_train = np.asfarray(loss_train, dtype=float) # converting to floats

# validation
file_val = open('/scratch/network/hishimwe/losses/val/val_30epochs','r')
loss_val = file_val.readlines()
# print(loss_val)
loss_val = np.array(loss_val) 
loss_val = np.asfarray(loss_val, dtype=float) # converting to floats

epochs = np.arange(len(loss_train)) # number of epochs run

plt.plot(epochs, loss_train, color="blue", label="Train Loss")
plt.plot(epochs, loss_val, color="green", label="Validation Loss")
plt.title("Training and Validation Losses")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

## Accuracy

```python
# training
file_train = open('/scratch/network/hishimwe/accuracies/train/train_30epochs','r')
ac_train = file_train.readlines()
ac_train = np.array(ac_train) 
ac_train = np.asfarray(ac_train, dtype=float) # converting to floats

# validation
file_val = open('/scratch/network/hishimwe/accuracies/val/val_30epochs','r')
ac_val = file_val.readlines()
# print(loss_val)
ac_val = np.array(ac_val) 
ac_val = np.asfarray(ac_val, dtype=float) # converting to floats

epochs = np.arange(len(ac_train)) # number of epochs run

plt.plot(epochs, ac_train, color="blue", label="Train Accuracy")
plt.plot(epochs, ac_val, color="green", label="Validation Accuracy")
plt.title("Training and Validation Accuracies")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

### Average Precision

```python
# training 
file_train = open('/scratch/network/hishimwe/precisions/train/train_30epochs','r')
prec_train = file_train.readlines()

num_epochs = len(prec_train)
avg_precisions_train = np.zeros((num_epochs,))

for i in range(num_epochs):
    l = np.array(prec_train[i].split(",")) # separate the values
    l = np.asfarray(l, dtype=float) # convert to floats
    avg_precisions_train[i] = np.average(l) # save average

    
# validation
file_val = open('/scratch/network/hishimwe/precisions/val/val_30epochs','r')
prec_val = file_val.readlines()

avg_precisions_val = np.zeros((num_epochs,))

for i in range(num_epochs):
    l = np.array(prec_val[i].split(",")) # separate the values
    l = np.asfarray(l, dtype=float) # convert to floats
    avg_precisions_val[i] = np.average(l) # save average
    
epochs = np.arange(num_epochs) # number of epochs run

plt.plot(epochs, avg_precisions_train, color="red", label="Train Average Precision")
plt.plot(epochs, avg_precisions_val, color="green", label="Validation Average Precision")
plt.title("Training and Validation Average Precisions")
plt.xlabel("Epochs")
plt.ylabel("Average Precisions")
plt.legend()
plt.show()
```
